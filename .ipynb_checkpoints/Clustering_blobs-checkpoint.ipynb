```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(random_state=33, n_samples=1000)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ward
agg = AgglomerativeClustering(n_clusters=4)
assignment = agg.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=assignment, s=60) 
#mglearn.discrete_scatter(X[:, 0], X[:, 1], assignment)
#plt.xlabel("Feature 0")
#plt.ylabel("Feature 1")
```

```python
#average
agg_a = AgglomerativeClustering(n_clusters=4 ,linkage='average')
assignment_a = agg_a.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=assignment_a, s=60) 
```

```python
#complete
agg_c = AgglomerativeClustering(n_clusters=4 ,linkage='complete')
assignment_c = agg_c.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=assignment_c, s=60) 
```

```python
k_means_0 = KMeans(n_clusters = 4)
k_means_0.fit(X)
```

```python
assignments = k_means_0.labels_
plt.scatter(X[:, 0], X[:, 1], c=assignments, s=60)
```

```python
from sklearn.cluster import DBSCAN
#X, y = make_blobs(random_state=0, n_samples=12)
dbscan = DBSCAN(eps=0.55)
clusters = dbscan.fit_predict(X)
print("Cluster memberships:\n{}".format(clusters))
```

```python
assignments_cl = clusters 
plt.scatter(X[:, 0], X[:, 1], c=assignments_cl, s=60)
```
